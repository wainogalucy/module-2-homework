### Cleaning and constructing Key Measurables

Injuries and robustness to destruction are a key interest point for the client. Clean and impute relevant columns and then create derived fields that best quantifies what the client wishes to track. **Use commenting or markdown to explain any cleaning assumptions as well as any derived columns you create.**

**Construct metric for fatal/serious injuries**

*Hint:* Estimate the total number of passengers on each flight. The likelihood of serious / fatal injury can be estimated as a fraction from this.

>Total.Fatal.Injuries<br>
>Total.Serious.Injuries<br>
>Total.Minor.Injuries<br>
>Total.Uninjured<br>

Then compute `fatal_injury_fraction` = (Fatal + Serious) / Total passengers.

```python
df= pd.read_csv("AviationData.csv", encoding="latin1:", low_memory= False)
```

```python
# Inspect injury columns
injury_column = ['Total.Fatal.Injuries', 'Total.Serious.Injuries',
               'Total.Minor.Injuries', 'Total.Uninjured']

print("Null counts in injury columns:")
print(df[injury_column].isnull().sum())
print("\nDescriptive statistics:")
print(df[injury_column].describe())

```

```python
# fill NaN values in injury columns with 0
# Rationale: if no injuries of a type were recorded, it is reasonable to treat as 0
for col in injury_column:
    df[col] = df[col].fillna(0)

# Estimate total number of people on board
df['total_Passengers'] = (df['Total.Fatal.Injuries'] +
                      df['Total.Serious.Injuries'] +
                      df['Total.Minor.Injuries'] +
                      df['Total.Uninjured'])

# Drop rows where no passenger info is available (total_aboard == 0)
df = df[df['total_Passengers'] > 0]
print(f"Rows after dropping zero-passenger records: {len(df)}")

# Create the key safety metric: fraction fatally or seriously injured
df['fatal_injury_fraction'] = (
    (df['Total.Fatal.Injuries'] + df['Total.Serious.Injuries']) / df['total_Passengers']
)

print("\nfatal_injury_fraction:")
print(df['fatal_injury_fraction'].describe())
```

### Investigate the *Make* column
- Identify cleaning tasks here
- List cleaning tasks clearly in markdown
- Execute the cleaning tasks
- For your analysis, keep Makes with a reasonable number (you can put the threshold at 50 though lower could work as well)

```python

df.info()
```

```python
#getting the first 40 rows
df["Make"].head(40)
```

```python
# Inspect and clean Make column and display the first 40 rows, including the null count
print("Null count:", df['Make'].isnull().sum())
print("\nTop 40 raw Make values:")
print(df['Make'].value_counts().head(40))
```

```python
# cleaning tasks:
#stripping off the whitespaces
#standardizing the cases
# Drop rows where Make is Null
df = df.dropna(subset=['Make'])

df['Make'] = (df['Make']
              .str.strip()
              .str.title()
         # Fix common multi-word make abbreviations
              .str.replace(r'\bCessna\b', 'Cessna', regex=True)
              .str.replace(r'\bPiper\b', 'Piper', regex=True)
              .str.replace(r'\bBeechcraft\b', 'Beech', regex=True)
              .str.replace(r'\bMcdonnell Douglas\b', 'McDonnell Douglas', regex=True)
              .str.replace(r'\bBoeing\b', 'Boeing', regex=True)
             )

# Count records per make
make_counts = df['Make'].value_counts()
print(f"Unique makes before threshold filter: {len(make_counts)}")

# Keep only makes with <= 50 records
valid_makes = make_counts[make_counts <= 50].index
df = df[df['Make'].isin(valid_makes)]

print(f"Unique makes after threshold filter (<=50 records): {df['Make'].nunique()}")
print(f"Rows retained: {len(df)}")
print("\nTop 40 makes by count:")
print(df['Make'].value_counts().head(40))
```

### Inspect Model column
- Get rid of any NaNs.
- Inspect the column and counts for each model/make. Are model labels unique to each make?
- If not, create a derived column that is a unique identifier for a given plane type.

```python
# Drop rows with missing model
df = df.dropna(subset=['Model'])
print(f"Rows after dropping null Model: {len(df)}")

# Check whether model names are unique to each make
print("\nTotal unique models:", df['Model'].nunique())

# Check if same model name appears under multiple makes
model_make_counts = df.groupby('Model')['Make'].nunique()
shared_models = model_make_counts[model_make_counts > 1]
print(f"\nModel names shared across multiple makes: {len(shared_models)}")
print("\nTop 10 shared model names:")
print(shared_models.sort_values(ascending=False).head(10))
```

```python
# Model labels are NOT unique to each make
# Create a make_model identifiers combining make and model
df['Model'] = df['Model'].str.strip().str.upper()
df['make_model'] = df['Make'].str.strip() + ' ' + df['Model'].str.strip()

print("Sample make_model values:")
print(df['make_model'].value_counts().head(20))
print(f"\nTotal unique make_model identifiers: {df['make_model'].nunique()}")
```

```python
# Engine.Type
print("Engine.Type values:")
print(df['Engine.Type'].value_counts())
print()

# Standardize casing and strip whitespace
df['Engine.Type'] = df['Engine.Type'].str.strip().str.title()

# Map 'Unknown' / 'None' to NaN
df['Engine.Type'] = df['Engine.Type'].replace({'Unknown': np.nan, 'None': np.nan, 'Nr': np.nan})

print("Engine.Type after cleaning:")
print(df['Engine.Type'].value_counts(dropna=False))
```

Column Removal

```python
# Re-calculate NaN percentages after cleaning
nan_pct = (df.isnull().sum() / len(df) * 100).sort_values(ascending=False)

print('NaN percentage per column (after cleaning):')
print(nan_pct.round(1))
```

```python
# Drop columns where more than 50% of rows are missing.
threshold = 50.0  # percent

cols_to_drop = nan_pct[nan_pct > threshold].index.tolist()
print(f'Columns to drop (>{threshold}% NaN):')
for c in cols_to_drop:
    print(f'  {c}: {nan_pct[c]:.1f}% missing')

df = df.drop(columns=cols_to_drop)
print()
print('Remaining columns:', df.shape[1])
print('Remaining rows:   ', df.shape[0])
print()
print('Columns kept:', df.columns.tolist())
```
